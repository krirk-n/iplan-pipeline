```python
mapid = "23917296"
wd = "D:/Projects/Epistemic Learning/iPlan/iplan-pipeline"

# Opening JSON file
d = open(wd + "/data/" + mapid + "_init_map.json") # generated from pipline
g = open(wd + "/data/" + mapid + "_reps.json") # generated from pipline #Name, Indicator, Threshold

init = json.load(d)
reps = json.load(g)

indicator_keys_mult_table = dict()

for i in range(len(reps)):
    indicator_keys_mult_table[reps[i]["Indicator"]] = dict()

for i in indicator_keys_mult_table.keys():
    for j in range(len(init["lucs"])):
        zone = init["lucs"][j]["zone"]
        indicator_keys_mult_table[i][zone] = init['lucs'][j]["multipliers"][i]
        
sign_multiplier_dict = dict()
insign_multiplier_dict = dict()
for i in indicator_keys_mult_table.keys():
    sign_multiplier_dict[i] = dict()
    insign_multiplier_dict[i] = dict()
```

```python
# Hand method
sign_multiplier_dict = dict()
insign_multiplier_dict = dict()
for i in indicator_keys_mult_table.keys():
    sign_multiplier_dict[i] = dict()
    insign_multiplier_dict[i] = dict()
for key in sign_multiplier_dict.keys():
    if key == "Birds":
        sign_multiplier_dict[key] = dict(filter(lambda elem: elem[1] > 60,indicator_keys_mult_table[key].items()))
        insign_multiplier_dict[key] = dict(filter(lambda elem: elem[1] <= 60,indicator_keys_mult_table[key].items()))
    if key == "GreenSpace":
        sign_multiplier_dict[key] = dict(filter(lambda elem: elem[1] > 0.6,indicator_keys_mult_table[key].items()))
        insign_multiplier_dict[key] = dict(filter(lambda elem: elem[1] <= 0.6,indicator_keys_mult_table[key].items()))
    if key == "LeadEmissions":
        sign_multiplier_dict[key] = dict(filter(lambda elem: elem[1] > 5,indicator_keys_mult_table[key].items()))
        insign_multiplier_dict[key] = dict(filter(lambda elem: elem[1] <= 5,indicator_keys_mult_table[key].items()))
    if key == "Butterflies":
        sign_multiplier_dict[key] = dict(filter(lambda elem: elem[1] > 1.4,indicator_keys_mult_table[key].items()))
        insign_multiplier_dict[key] = dict(filter(lambda elem: elem[1] <= 1.4,indicator_keys_mult_table[key].items()))
    if key == "GHG":
        sign_multiplier_dict[key] = dict(filter(lambda elem: elem[1] > 200,indicator_keys_mult_table[key].items()))
        insign_multiplier_dict[key] = dict(filter(lambda elem: elem[1] <= 200,indicator_keys_mult_table[key].items()))   
    if key == "Heat":
        sign_multiplier_dict[key] = dict(filter(lambda elem: elem[1] > 2.9,indicator_keys_mult_table[key].items()))
        insign_multiplier_dict[key] = dict(filter(lambda elem: elem[1] <= 2.9,indicator_keys_mult_table[key].items()))
    if key == "Housing":
        sign_multiplier_dict[key] = dict(filter(lambda elem: elem[1] > 1,indicator_keys_mult_table[key].items()))
        insign_multiplier_dict[key] = dict(filter(lambda elem: elem[1] <= 1,indicator_keys_mult_table[key].items()))
    if key == "Population":
        sign_multiplier_dict[key] = dict(filter(lambda elem: elem[1] > 4,indicator_keys_mult_table[key].items()))
        insign_multiplier_dict[key] = dict(filter(lambda elem: elem[1] <= 4,indicator_keys_mult_table[key].items()))
    if key == "ISA":
        sign_multiplier_dict[key] = dict(filter(lambda elem: elem[1] > 19,indicator_keys_mult_table[key].items()))
        insign_multiplier_dict[key] = dict(filter(lambda elem: elem[1] <= 19,indicator_keys_mult_table[key].items()))
```

```python
# MAD method    
for key in indicator_keys_mult_table.keys():
    values = np.array(list(indicator_keys_mult_table[key].values()))
    median = np.median(values)
    mad = np.median(np.abs(values - median))
    threshold_multiplier = 100 # you can change it as appropriate
    threshold_value = median + (threshold_multiplier * mad)
    sign_multiplier_dict[key] = dict(filter(lambda elem: elem[1] >= threshold_value, indicator_keys_mult_table[key].items()))
    insign_multiplier_dict[key] = dict(filter(lambda elem: elem[1] < threshold_value, indicator_keys_mult_table[key].items()))
    if len(insign_multiplier_dict[key]) == 0: # handle empty dict
        min_value = min(indicator_keys_mult_table[key].values())
        sign_multiplier_dict[key] = dict(filter(lambda elem: elem[1] > min_value, indicator_keys_mult_table[key].items()))
        insign_multiplier_dict[key] = dict(filter(lambda elem: elem[1] == min_value, indicator_keys_mult_table[key].items()))
    elif len(sign_multiplier_dict[key]) == 0: # handle empty dict
        max_value = max(indicator_keys_mult_table[key].values())
        insign_multiplier_dict[key] = dict(filter(lambda elem: elem[1] < max_value, indicator_keys_mult_table[key].items()))
        sign_multiplier_dict[key] = dict(filter(lambda elem: elem[1] == max_value, indicator_keys_mult_table[key].items()))
```
