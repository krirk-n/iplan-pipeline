```python
from pyArango.connection import *
import pandas as pd
import matplotlib.pyplot as plt
from cycler import cycler
```

```python
db = conn["lem_user_data"]
```

```python
aql = '''FOR s in submissions_prod
        FILTER s.mapKey == "57718789"
        RETURN s
        '''
queryResult = db.AQLQuery(aql, rawResults=True, batchSize=100)
queryResult
```

```python
# Define column names
columns = ['userKey', 'submissionCount', 'approvalCount', 'date']

# Create an empty DataFrame with these columns
df = pd.DataFrame(columns=columns)

for submission in queryResult:
    df.loc[len(df.index)] = [submission['userKey'], submission['submissionCount'], submission['approvalCount'], submission['date']]

df = df.sort_values(by=['userKey', 'submissionCount']).reset_index(drop=True)
df['interval_sec'] = 0
for i in range(1, len(df)):
    if df.iloc[i]['userKey'] == df.iloc[i - 1]['userKey']:
        df.loc[i, 'interval_sec'] = df.iloc[i]['date'] - df.iloc[i - 1]['date']
    else:
        df.loc[i, 'interval_sec'] = 0

df['interval_sec'] = df['interval_sec'] / 1000
df
```

```python
linestyle_cycler = cycler('linestyle',['-','--',':','-.'])
```

```python
plt.figure(figsize=(10, 6))
for label, group in df.groupby('userKey'):
    linestyle_cycler = cycler('linestyle',['-','--',':','-.'])
    plt.rc('axes', prop_cycle=linestyle_cycler)
    plt.plot(group['submissionCount'], group['approvalCount'], label=label)

plt.xlabel('submissionCount')
plt.ylabel('approvalCount')
plt.title('Approval count for each submission')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
for label, group in df.groupby('userKey'):
    plt.plot(group['submissionCount'], group['interval_sec'], label=label)

plt.xlabel('submissionCount')
plt.ylabel('interval_sec')
plt.title('Time spent before each submission')
plt.legend()
plt.show()
```
